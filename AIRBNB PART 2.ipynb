
# Stage 2: Parameter Optimization for SARIMAX Model
In the process of building a SARIMAX (Seasonal AutoRegressive Integrated Moving Averages with eXogenous regressors) model, one of the most crucial steps is determining the optimal parameters for our model. These parameters include the order of AR, I, and MA (p, d, q), and the seasonal order (P, D, Q, m).

## Why Do We Need Parameter Optimization?
Improved Model Performance: A model with the correct parameters will be better able to capture the time series patterns, thereby leading to improved forecasting performance.

#### Avoiding Overfitting and Underfitting:
If our model is too complex (high orders of p, d, q), we might overfit to the training data and perform poorly on unseen data. Conversely, a model that is too simple (low orders of p, d, q) might not capture the essential structures in the data, leading to an underfitting model.

#### Efficient Resource Usage: 
Models with unnecessary complexity (higher order terms that are not needed) may consume more computational resources without leading to significant improvements in forecast accuracy.

#### How Do We Conduct Parameter Optimization?
To determine the best parameters for our SARIMAX model, we use a method called "Grid Search". It's an exhaustive search over a specified parameter space. In our case, we consider p, d, and q values from 0 to 2, both for regular and seasonal parts of the model, which gives us a total of 333*3 = 81 combinations of parameters to evaluate.

For each combination, we fit a new SARIMAX model and calculate the Akaike Information Criterion (AIC), a metric that quantifies the goodness of fit of our model and the complexity of the model. The lower the AIC, the better the model. Therefore, we look for the combination of parameters that result in the lowest AIC.

##### The SARIMAX model we fit includes both an endogenous variable ('total', representing the total number of calls), and several exogenous variables representing factors we believe might influence the total number of calls (e.g., 'english', 'german', 'french', 'italian', 'spanish', and 'covid_period'). The "seasonality" in SARIMAX is accounted for by including a seasonal order (P,D,Q,m) in our model, with m being 54 (corresponding to 54 weeks in a year).


```python
from pmdarima import auto_arima
import itertools
import warnings
import statsmodels.api as sm

# define the p, d, q values to take any value between 0 and 2
p = d = q = range(0, 2)

# generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q))

# generate all different combinations of seasonal p, q and q triplets (use 54 for weekly seasonality)
seasonal_pdq = [(x[0], x[1], x[2], 54) for x in list(itertools.product(p, d, q))]

# specify to ignore warning messages
warnings.filterwarnings("ignore")

# Grid Search to find the optimal parameters
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(data['total'], 
                                exog=data[['english', 'german', 'french', 'italian', 'spanish', 'year', 'month', 'day', 'weekday', 'covid_period']], 
                                order=param,
                                seasonal_order=param_seasonal,
                                enforce_stationarity=False,
                                enforce_invertibility=False)

            results = mod.fit()

            print('ARIMA{}x{} - AIC:{}'.format(param, param_seasonal, results.aic))
        except Exception as e:
            print(f'Exception {e} occurred for ARIMA{param}x{param_seasonal}.')
            continue

```

```python
from pmdarima import auto_arima

# suppress warnings
warnings.filterwarnings("ignore")

exog_vars = data[['english', 'german', 'french', 'italian', 'spanish', 'year', 'month', 'day', 'weekday', 'covid_period']]

# define auto_arima model
model = auto_arima(y=data['total'], 
                   exogenous=exog_vars, 
                   start_p=0, start_q=0,
                   test='adf',       # use adftest to find optimal 'd'
                   max_p=2, max_q=2, # maximum p and q
                   m=54,             # frequency of series
                   d=None,           # let model determine 'd'
                   seasonal=True,    # No Seasonality
                   start_P=0, 
                   D=0, 
                   trace=True,
                   error_action='ignore',  
                   suppress_warnings=True, 
                   stepwise=True)

print(model.summary())
```
